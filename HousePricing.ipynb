```r
num_vars <- subset(data[sapply(data, is.numeric)], select= -c(Bedrooms, Bathrooms))

melted <- melt(num_vars)
```

```r
library(ggplot2)
library(reshape2)
```

```r
ggplot(melted, aes(x = value)) +
  geom_histogram(bins = 30, fill = "steelblue", color = "white") +
  facet_wrap(~variable, scales="free", ncol = 4) +
  theme(aspect.ratio = 1)  # deixa cada facet quadrado
```
